## Exercises

This homework aims to analyse data generated by political candidates during the Irish General Elections 2016.
We have a CSV file describing candidates (specific to each student) and a json file (common to all students) describing Twitter activity of candidates.
The aim is to merge the 2 files into a single CSV file and produce a data quality report and a data quality plan for the merged CSV, as well as some initial interpretation of findings. Each student will get a different subset of candidates. The CSV file is named using the format: **ge16-results-elected-111111.csv**, where the number 111111 stands for the student number. You need to work with the CSV file corresponding to your student number. It contains data describing the candidates, e.g., name, party and constituency, and data on whether they got elected or not. The json file contains data such as the candidate name and the number of tweets issued by their Twitter account, the number of followers the Twitter account has, etc. We aim to understand this data and to look for initial evidence that candidate Twitter account details (e.g., tweets, folowers) is (or not) indicative of getting elected.
