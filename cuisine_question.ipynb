## Q. Does the type of cuisine a business offers affect its success?
In answering this question, we aim to discover if a certain type of cuisine could be recommended to a new business owner in order to increase their likelyhood of being successful.

A restaurant would want to maximise their positive reviews, thereby increasing the likelyhood that more potential customers will see the reviews and visit the restaurant. We hope that a predictive analytics model will be able to tell how likely a business is to be successful based on the cuisine(s) that they offer, and as a result, provide information on which cuisine(s) a restaurant should offer to increase ratings.

### Further Data Quality Plan
Some issues we may face in answering the question with this data set:
1. Duplicate rows as a result of multiple reviews for each business. This could skew the data.
2. The categories of businesses are strings, not lists.
3. Possibility that not all businesses will have a particular cuisine associated with them.
4. Difficult to work with feature of list of categories, very few will have the same list.
5. There may be some businesses with too few reviews, meaning inaccurate results.

Suggested solutions:
1. Remove the duplicate rows, only keeping one row per business.
2. Convert these categories to lists in order to make them more workable for our purposes.
3. If any businesses do not have specific cuisine(s) associated, remove them from the dataframe.
4. Convert single feature of lists of categories into many binary numeric features, one for each category, or if each of the lists of categories only have one real cuisine (i.e. not 'Restaurants' or 'Food' etc.), keep only the significant category.
5. Remove businesses with too few businesses

###### But first, we will convert these categories strings into lists in order to make them more workable for our purposes

```python
# show type of object in each cell in categories column
print(smdf.categories.iloc[1]) # showing contents of a cell
type(smdf.categories.iloc[1]) # showing type of data in cell
```

```python
# contents of cells look like lists but are actually strings
# create function to convert string to array
def string_to_array(string):
    """Converts string in cell to array that can be indexed"""
    string = string.replace(" ", "")
    string = string.replace("[", "")
    string = string.replace("]", "")
    string = string.replace(" ", "")
    string = string.replace("'", "")
    string = string.split(',')
    return string
```

```python
# list of cuisine types - used to determine if a restaurant/business has a particular cuisine associated
cuisine_types = ['American(Traditional)', 'Burgers', 'FastFood', 'American(New)',
                    'Pizza', 'Gluten-Free', 'AsianFusion', 'Chinese', 'Diners', 'Seafood', 'Japanese',
                    'Tapas/SmallPlates', 'Barbeque', 'Mediterranean', 'Vegetarian', 'Portuguese', 'German',
                    'SushiBars', 'Mexican', 'Tex-Mex', 'ChickenWings', 'Steakhouses',
                    'Korean', 'Greek', 'Irish', 'Indian', 'Vietnamese', 'Thai', 'Peruvian', 'LatinAmerican',
                    'Buffets', 'Cajun/Creole', 'SoulFood', 'Southern', 'French', 'Caribbean', 'Fondue', 'Taiwanese',
                    'Lebanese', 'MiddleEastern', 'Spanish', 'EthnicFood', 'Fish&Chips', 'Salvadoran', 'British',
                    'Kosher', 'Breakfast&Brunch', 'Sandwiches', 'Italian']

# to count number of businesses with no specified cuisine
no_cuisine_count = 0
# list keeps track of if business has cuisine or not
has_cuisine_list = []
# list to hold the category features converted from strings to lists
row_array_list = []

# loop to determine if business has a specified cuisine
for i in range(len(smdf.categories)):
    has_cuisine = False
    row = string_to_array(smdf.categories.iloc[i]) # converting string of categories into list to iterate over
    for j in row: # for each category associated with the row
        if j in cuisine_types: # if one of the assiciated categories is a cuisine listed in cuisine_types
            has_cuisine = True
    if has_cuisine:
        has_cuisine_list.append(1)
    else:
        no_cuisine_count +=1
        has_cuisine_list.append(0)
    row_array_list.append(row)

# print length of smdf and number of businesses with no associated cuisine to
# determine if dataset will be large enough

print(len(smdf))
print('Number of businesses with no specific cuisine:', no_cuisine_count)
```

19 of 352 businesses have no associated cuisine. This leaves a remaining 333 businesses with a particular cuisine/combination of cuisines. This amount is reasonable to use for creating a predictive model.

```python
# append row_array_list to dataframe
smdf['list_categories'] = row_array_list

# append has_cuisine_list to dataframe
smdf['has_cuisine'] = has_cuisine_list

# show head of dataframe - can see that categories and list_categories appear to have same content
smdf.head(10)
```

## Predictive Modeling - Linear Regression
### Train a linear regression model to predict the target feature 'Business stars', using the categories within the 'categories' feature

```python
# Training the data
lm_categ = sm.ols(formula="Business_stars ~ american_trad + burgers + fast_food + american_new + pizza + \
gluten_free + asian_fusion + chinese + diners + seafood + japanese + tapas_smallplates + barbeque + mediterranean +\
vegetarian + portuguese + german + sushi_bars + mexican + tex_mex + chickenwings + steakhouses + korean + greek +\
irish + indian + vietnamese + thai + peruvian + latin_american + buffets + cajun_creole + soulfood + southern +\
french + caribbean + fondue + taiwanese + lebanese + middle_eastern + spanish + ethnic_food + fish_chips + salvadoran +\
british + kosher + breakfast_brunch + sandwiches + italian", data=smdf).fit()
```

```python
print(lm_categ.summary())
```

#### Statistical Significance of Coefficients
We can see that only 4 of the 49 categories of restaurants have a confidence interval that does not contain 0 (fast_food, diners, chickenwings, and cajun_creole), along with the intercept. As a result, we reject the null hypothesis for these categories. We can't reject the null hypothesis for the rest of the categories, which we can also see have p-values greater than 0.05 - we now know that these categories are statistically insignificant.

For these 4 categories of restaurants that have confidence intervals not containing 0, we see that their p-values are below 0.05: fast_food's p-value is 0.001, diners' p-value is 0.04, chickenwings' p-value is 0, and cajun_creole's p-value is 0.035. These p-values show that the coefficients of these categories are statistically significant.

Conclusion: While most of the categories have been shown to be statistically insignificant, we cannot remove the categories with statistically insignificant weights because all of these categories are part of the same 'categories' feature. Removing the statistically insignificant categories would skew our data to be less accurate. We have seen that four of the categories are statistically significant, and as a result, we will test the accuracy of the model on our dataset.

### Using the above trained model, print the predicted target feature for all the examples in the training set. Threshold the predicted target feature to get the predicted class for each example. Print the accuracy of the model on the training set

```python
# print predicted target feature for all examples in the training set
lm_categ.predict(smdf)
```

###### Threshold the predicted target feature to get the predicted class for each example.

```python
# We threshold at the values that are between the two class values, e.g. threshold at 4.75 to decide if a predicted 
# business stars rating should be 5 stars or below 5 stars. The further conditional statements to find the
# correct class.

# add column for values of prediction of target feature
smdf['predicted_business_stars'] = lm_categ.predict(smdf)

# list to hold thresholded predictions
thresholded_rating = []

# loop to determine if business has a specified cuisine
for i in range(len(smdf)):
    if smdf.predicted_business_stars.iloc[i] > 4.75:
        thresholded_rating.append(5.0)
    elif smdf.predicted_business_stars.iloc[i] > 4.25:
        thresholded_rating.append(4.5)
    elif smdf.predicted_business_stars.iloc[i] > 3.75:
        thresholded_rating.append(4.0)
    elif smdf.predicted_business_stars.iloc[i] > 3.25:
        thresholded_rating.append(3.5)
    elif smdf.predicted_business_stars.iloc[i] > 2.75:
        thresholded_rating.append(3.0)
    elif smdf.predicted_business_stars.iloc[i] > 2.25:
        thresholded_rating.append(2.5)
    elif smdf.predicted_business_stars.iloc[i] > 1.75:
        thresholded_rating.append(2.0)
    elif smdf.predicted_business_stars.iloc[i] > 1.25:
        thresholded_rating.append(1.5)
    elif smdf.predicted_business_stars.iloc[i] > 0.75:
        thresholded_rating.append(1.0)
    elif smdf.predicted_business_stars.iloc[i] > 0.25:
        thresholded_rating.append(0.5)
    else:
        thresholded_rating.append(0.0)

# change predicted_business_stars column to hold thresholded values
smdf['predicted_business_stars'] = thresholded_rating
```

```python
# count correct predictions and number of incorrect predictions for each level of incorrect prediction

# initialise counter variables to count the number of correct predictions, and the number of incorrect predictions for
# each error level
correct = 0
error_0_5 = 0
error_1 = 0
error_1_5 = 0
error_2 = 0
error_2_5 = 0
error_3 = 0
error_3_5 = 0
error_4 = 0
error_4_5 = 0
error_5 = 0
for i in range(len(smdf)):
    x = abs(smdf.Business_stars.iloc[i] - smdf.predicted_business_stars.iloc[i])
    if x == 0:
        correct += 1
    elif x == 0.5:
        error_0_5 += 1
    elif x == 1.0:
        error_1 += 1
    elif x == 1.5:
        error_1_5 += 1
    elif x == 2.0:
        error_2 += 1
    elif x == 2.5:
        error_2_5 += 1
    elif x == 3.0:
        error_3 += 1
    elif x == 3.5:
        error_3_5 += 1
    elif x == 4.0:
        error_4 += 1
    elif x == 4.5:
        error_4_5 += 1
    elif x == 5.0:
        error_5 += 1
        
print('Correct predictions:', correct, '\nError of 0.5:', error_0_5, '\nError of 1.0:', error_1, 
      '\nError of 1.5:', error_1_5,
      '\nError of 2.0:', error_2, 
      '\nError of 2.5:', error_2_5,
      '\nError of 3.0:', error_3,
      '\nError of 3.5:', error_3_5,
      '\nError of 4.0:', error_4,
      '\nError of 4.5:', error_4_5,
      '\nError of 5.0:', error_5)
```

## Evaluation/Conclusion
We have shown in the above calculations that while the prediction model we trained was more accurate than a very simple baseline prediction model that used the average number of business stars as the prediction, our model was not accurate enough to be used as a reliable predictor of a business's rating. With an accuracy of just 39%, it would be incorrect more often than not.

This model was trained on the entire data set. If we had a better accuracy, the next step would have been to train the model on a training subset of the data and then test it on the remaining data, and then test its performance. As we know, more complex models (e.g. with more features) can fit the training data better, but do not generalise to new data, since they pick up too many unimportant details from the training data. The more complex the model, the better the accuracy on the training data, but the accuracy on the test data will be worse. As this model was trained with a lot of 'features' (the columns representing the types of restaurant within the 'categories' feature), we know that if we were to train the model on a subset of this data and then measure the accuracy on a test subset, the accuracy would be even lower.

Rescaling the descriptive features is not possible on our data, because the features are already binary numeric features. The removal of statistically insignificant categories in our data would be incorrect as they are part of the same overall feature ('categories') as the statistically significant categories.

Taking the above information into account, we have arrived at an answer to our initial question of 'does the type of cuisine a business offers affect its success?'. We have found that, based on the data provided by Yelp, the answer is no - there is not enough evidence that the type of cuisine a business offers affects its success in any very meaningful way. From the model created, it would not be possible to provide a business with a reliable recommendation of what cuisine they should offer to maximise their ratings.
